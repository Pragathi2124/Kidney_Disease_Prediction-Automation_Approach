```python
import numpy as np
import pandas as pd
dataset = pd.read_csv("/content/cleaned_dataset.csv").drop("Unnamed: 0", axis=1)
dataset.head()
```

```python
dataset.columns
```

```python
training = dataset.sample(frac=0.8, random_state=42).reset_index(drop=True)
test = dataset.drop(training.index).reset_index(drop=True)
```

# Automation method - pycaret package/library

```python
#!pip install pycaret
```

```python
from pycaret.classification import *
```

```python
exp_classification = setup(data=training, target='classification', session_id=123)
```

```python
compare_models()
```

```python
et = create_model('et')
```

```python
# Fine Tuning
tuned_et = tune_model(et)
```

```python
# Evaluate the model
evaluate_model(et)
```

```python
# predict the unseen dataset
unseen_prediction = predict_model(et, data=test)
unseen_prediction.head()
```

```python
# save model
save_model(et, "et_model")
```
